```python
# median of the array
arr=[1,2,3,4,5]
n=len(arr)
if n%2==0:
  median=(arr[n//2-1] + arr[n//2]/2)
else:
  median=arr[n//2]
print("The median of the array is:", median)
```

```python
# mode of the array
arr=[10,20,30,40,50]
f={}
for num in arr:
  if num in f:
    f[num]+=1
  else:
    f[num]=1
mode=max(f, key=f.get)
print("The mode of the array is:", mode)
```

```python
# sum of array
arr=[1,2,3,4,5]
print(arr)
sum_cal=0
for i in arr:
  sum_cal=sum_cal+i
print("the sum of the array element is:",sum_cal)
```

```python
#mean
arr=[1,2,3,4,5]
print(arr)
sum_cal=0
for i in arr:
  sum_cal=sum_cal+i
print("the sum of the array element is:",sum_cal)
mean=sum_cal/len(arr)
print("mean:",mean)
```

```python

import numpy as np
speed = [100,99,34,56,34,8,39,23,12]
x = np.mean(speed)
print(x)
```

```python
import numpy as np

speed = [100,99,34,56,34,8,39,23,12]

# Calculate the 75th percentile
p_75 = np.percentile(speed, 75)

print(f"The original speed array: {speed}")
print(f"The 75th percentile of the speed array is: {p_75}")
print("This means that 75% of the speeds are less than or equal to {p_75}.")

# For comparison, the median is the 50th percentile
p_50 = np.percentile(speed, 50)
print(f"The 50th percentile (median) is: {p_50}")
```

```python
# for mode we have another lib called scipy
from scipy import stats
speed = [100,99,34,56,34,8,39,23,12]
z=stats.mode(speed)
print(z)
```

```python
# for standard deviation
import numpy as np
speed=[12,45,34,76,45,89,76]
x=np.std(speed)
print(x)
```

```python
# for variance
import numpy as np
speed=[12,45,34,76,45,89,76]
x=np.var(speed)
print(x)
```

```python
# for percentile
import numpy as np
speed=[12,45,34,76,45,89,76]
x=np.percentile(speed,75)
print(x)
```
