```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
m=numerator/denominator
b=y_mean-m*x_mean
print("slope : ",m)
print("intercept : ",b)
```

```python
# prediction
y_pred=m*x +b
print("predictiopn : ",y_pred)
```

```python
error = y-y_pred
print("error : ",error)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt
```

```python
# using model to do linear regression
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
x=np.array([[0],[1],[2],[3],[4]])
y=np.array([2,3,4,5,6])
model = LinearRegression()
model.fit(x,y)
```

```python
print("slope : ",model.coef_[0])
print("intercept : ",model.intercept_)
predictions = model.predict(x)
print("predictions : ",predictions)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
# calculating the errors in lineear regression
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
# lets take a sample data
y_true=[3,-0.5,2,7]
y_pred = [2.5,0.0,2,8]
# calculate the matrix
mae = mean_absolute_error(y_true,y_pred)
mse = mean_squared_error(y_true,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true,y_pred)
# printing the results
print(f" Mean absoluter error (MAE) : {mae}")
print(f" Mean squared error (MSE) : {mse}")
print(f" Root mean squared error (RMSE) : {rmse}")
print(f"r-squared (R2) : {r2}")
```

```python
# k fold cross validation on mlr
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
print(df)
```

```python
x = df[['age', 'experience']]
y = df['income']
model = LinearRegression()
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, x, y, cv=kf, scoring='r2')
print("R2 score for each fold:", scores)
print("Average r2 score", scores.mean)
```

```python
mse_scores = cross_val_score(model, x, y, cv=kf, scoring='neg_mean_squared_error')
print("MSE for each fold:", mse_scores)
print("Average MSE:", -mse_scores.mean())
rmse = np.sqrt(-mse_scores.mean())
print("RMSE:", rmse)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
rmse_list = []
r2_list=[]
fold_numbers=[]
fold=1
for train_index, test_index in kf.split(x):
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model = LinearRegression()
  model.fit(X_train, y_train)

  y_pred = model.predict(X_test)

  rmse = np.sqrt(np.mean((y_test - y_pred) ** 2))
  rmse_list.append(rmse)

  r2 = model.score(X_test, y_test)
  r2_list.append(r2)

  fold_numbers.append(fold)
  print(f"Fold {fold} --> RMSE: {rmse:.2f}, R²: {r2:.2f}")
  fold += 1


print("RMSE for each fold:", rmse_list)
print("Average RMSE:", np.mean(rmse_list))
```

```python
# viualization
plt.figure(figsize=(12,5))

# RMSE plot
plt.subplot(1,2,1)
plt.plot(fold_numbers, rmse_list, marker='o', color='red')
plt.title('RMSE per Fold')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.grid(True)

# R² plot
plt.subplot(1,2,2)
plt.plot(fold_numbers, r2_list, marker='o', color='green')
plt.title('R² per Fold')
plt.xlabel('Fold')
plt.ylabel('R²')
plt.grid(True)

plt.tight_layout()
plt.show()
```
