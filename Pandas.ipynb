```python
d={'one':pd.Series([1,2,3], index=['a','b','c']),
   'two':pd.Series([1,2,3,4], index=['a','b','c','d'])}
D1=pd.DataFrame(d)
print(D1)
```

```python
# ADDING A COLUMN BY PASSINGSERIES
d['three']=pd.Series([10,20,30], index=['a','b','c'])
print(d)
```

```python

#ADDING A COLUMN IN THE DATAFRAME USNG THE EXISTING COLUMN
d['five']=d['one']+d['three']
print(d)
```

```python
# slicing in dataframe
d={'one':pd.Series([1,2,3], index=['a','b','c']),
   'two':pd.Series([1,2,3,4], index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df[2:4])
```

```python
# addition of rows
d = {
    'one': pd.Series([1, 2, 3], index=['a', 'b', 'c']),
    'two': pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
}

df = pd.DataFrame(d)
print(df)
```

```python
df2 = pd.DataFrame([[5, 6], [7, 8]], columns=['one', 'two'])

df = pd.concat([df, df2])
print(df)
```
